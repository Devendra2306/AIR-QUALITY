```python
import duckdb as ddb
```

```python
ddb.sql("INSTALL https; LOAD https")
```

```python
con = ddb.connect("../air_quality.db")
```

```python
con.execute("CREATE schema IF NOT EXISTS raw")
```

```python
con.sql("""
SET s3_access_key_id='';
SET s3_secret_access_key='';
SET s3_region='';
""")
```

```python
con.execute("""
CREATE TABLE IF NOT EXISTS raw.air_quality_data (
    location_id BIGINT,
    sensors_id BIGINT,
    "location" VARCHAR,
    "datetime" TIMESTAMP,
    lat DOUBLE,
    lon DOUBLE,
    "parameter" VARCHAR,
    units VARCHAR,
    "value" DOUBLE,
    "month" VARCHAR,
    "year" BIGINT,
    ingestion_datetime TIMESTAMP
);
""")
```

```python
con.execute("""
INSERT INTO raw.air_quality_data
SELECT
    location_id,
    sensors_id,
    "location",
    "datetime",
    lat,
    lon,
    "parameter",
    units,
    "value",
    "month",
    "year",
    current_timestamp AS ingestion_datetime
FROM read_csv('s3://openaq-data-archive/records/csv.gz/locationid=9802/year=2025/month=01/*.csv.gz');
""")
```

```python
con.close()
```
