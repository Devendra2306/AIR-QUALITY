```python
import json

from openaq import OpenAQ
```

```python
import json

with open("../secrets.json") as f:
    secrets = json.load(f)

openaq_key = secrets["openaq-api-key"]
```

```python
client =OpenAQ(api_key=openaq_key)

client
```

```python
mu_bbox_search_location = client.locations.list(
    bbox=(-80.349613,40.180593,-71.582523,43.328498),
    limit=1000
)

mu_bbox_search_location.meta
 
```

```python
mu_bbox_search_location.results
```

```python
locations_info = {}

for location in mu_bbox_search_location.results:
    locations_info[location.id] = location.name

locations_info
```

```python
len(locations_info)
```

```python
import json

with open("../location.json", "w", encoding="utf-8") as f:
    json.dump(locations_info, f, ensure_ascii=False, indent=4)
```
